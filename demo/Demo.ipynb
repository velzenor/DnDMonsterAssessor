```python
#your custom monster's stats
armor_class = 12
hit_points = 10
hit_dice = None
multiattack = False
legendary_resistances = False
legendary_actions = False
flying = True
strength = 7
dexterity = 15
constitution = 10
intelligence = 8
wisdom = 8
charisma = 8
```

```python
import pickle

with open('trained_model.pkl','rb') as file:
  model = pickle.load(file)
```

```python
import math

#defaults
data = {'AC':armor_class, 'HP':hit_points, 'HitDice':hit_dice, 'Multiattack':multiattack, \
        'LegRes':legendary_resistances, 'LegAct':legendary_actions, 'Flying':flying, \
        'Str':strength, 'Dex':dexterity, 'Con':constitution, \
        'Int':intelligence, 'Wis':wisdom, 'Cha':charisma}

for score in ['Str','Dex','Con','Int','Wis','Cha']:
  if not data[score]: data[score] = 10

if not armor_class:
  data['AC'] = 10 + data['Dex']

if hit_points and not hit_dice:
  data['HitDice'] = int(data['HP']/(5.5+math.floor((constitution-10)/2)))
```

```python
import pandas

cr_estimate_float = model.predict(pandas.DataFrame([data]))[0]
```

```python
for denom in ['','/2','/4','/8']:
  if round(cr_estimate_float)>=1:
    cr_estimate = str(round(cr_estimate_float))+denom
    break
  cr_estimate_float *= 2
else:
  cr_estimate = '0'
```

```python
#your custom monster's CR estimate
print(cr_estimate)
```
